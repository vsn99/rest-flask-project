```python
import pandas as pd
import requests
```

```python
response = requests.get("https://pokeapi.co/api/v2/pokemon/?limit=51&offset=0")
response
```

```python
data = response.json()
data
```

```python
data = data.get("results")
data
```

```python
for record in data:
    url = record.get("url")
    response = requests.get(url)
    individualData = response.json()
    record["weight"] = individualData.get("weight")
    types = individualData.get("types")[0]
    types = types.get("type")
    record["element"] = types.get("name")
    print(record)
```

```python
df = pd.DataFrame(data)
df
```

```python
# Show a column
df['url']
```

```python
# 	Remove columns or rows	Drop unnecessary columns

df = df.drop("url", axis=1)
```

```python
avg_weight_by_type = df.groupby('element')['weight'].mean()
avg_weight_by_type
```

```python
def weight_category(w):
    if w < 50:
        return 'Light'
    elif w < 150:
        return 'Medium'
    else:
        return 'Heavy'

df['weight_category'] = df['weight'].apply(weight_category)
```
